bec , token ID only convert the text -> integers but its loos the meaning of the different text . 
like - dog = 303 token ID 
     - cat = 444 token ID 
     both are animal but this token ID do not capture that meaning

### 1. Hands on with token Embedding

```python
import torch
input_token_id = torch.tensor([2,4,1])
```

```python
vocab_size = 5
output_dim=3
torch.manual_seed(0)
```

```python
embedding_layer = torch.nn.Embedding(vocab_size, output_dim)
print("Embedding layer weights:", embedding_layer.weight)
embedded_output = embedding_layer(input_token_id)
embedded_output.shape
```

### 2. Hands on token embedding models


```python
## import training model
import gensim.downloader as api
model = api.load("glove-wiki-gigaword-50")
```

```python
word_vector=model
print("Vector representation of the word 'work':", word_vector['work'])
```

```python
print(word_vector['work'].shape)
```

## King + women - man 

```python
print(word_vector.most_similar(positive=['king','women'], negative=['man'],topn=10))
```

this model is not good to capture the meaning of the word . this is light weight 60 MB model . 

for working on real peojects use the google and other model parameters and dim model 

```python
## similarty between two words
print(word_vector.similarity('woman','man'))
print(word_vector.similarity('woman','queen'))
print(word_vector.similarity('woman','girl'))
print(word_vector.similarity('woman','boy'))
print(word_vector.similarity('woman','aunty'))
print(word_vector.similarity('woman','computer'))
```

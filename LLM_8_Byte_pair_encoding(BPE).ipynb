## Byte pair encoding 

it is the middle solution of both word based tokenization and character basef tokenization


#### Rule -1 Do not split frequently words into smaller sub-words

#### Rule -2 split rare words into the smaller words or sub-words of in characters if needed

implementation of BPE is given in the tiktoken library 
it handle the algo of the BPE using RUST language

```python
import importlib
import tiktoken

print("version of the tiktoken = ", importlib.metadata.version("tiktoken"))
```

```python
## instance of the tiktoken
tokenizer = tiktoken.get_encoding("gpt2")
```

```python
text = ("How are you aditye? <|endoftext|> what about the work you are doing ")

integers = tokenizer.encode(text,allowed_special={"<|endoftext|>"})
print(integers)
```

```python
# how can i check the last token of the tiktoken and also <|endoftext|>
tokenizer.encode("<|endoftext|>",allowed_special={"<|endoftext|>"})
```

```python
# convert token IDs into the token
words=tokenizer.decode(integers)
print(words)
```

```python
# how the BPE handeles the unknown words lets check
unknown=tokenizer.encode("Akebswi#4jsn") 
print(unknown)
print(tokenizer.decode(unknown))
```
